```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load data & initial checks

df=pd.read_csv("weather_data.csv") # Read CSV file
print(df)
print(df.head())
print(df.info())
print(df.describe())
print(df.isnull().sum())
print(df.duplicated())
```

```python
# Data cleaning

df['Date']=pd.to_datetime(df['Date'])  #Convert date int type to date type
print(df)

df.to_csv('weather_data_clean.csv', index=False)
print('Saved')
```

```python
#Basic Exploratory Data Analysis (EDA)

Statistics = df[['Temperature', 'Humidity','WindSpeed','Rainfall']].agg(['mean', 'median', 'min', 'max'])  #Using agg() find statistical operations like mean,median,min,max,etc.
print('\nMean,Median,Min,Max of all numeric column are :')
print(Statistics)

print('Hottest day :', df.loc[df['Temperature'].idxmax()]) #Print all information of Hottest day
print('Coldest day :', df.loc[df['Temperature'].idxmin()]) #Print all information of Coldest day
```

```python
#Visualization


plt.plot(df.Date,df.Temperature)
plt.title('Line plot')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.show()

plt.hist(df.Temperature,bins=30,color='skyblue')
plt.title('Histogram')
plt.xlabel('Temperature')
plt.ylabel('Frequency')
plt.show()

plt.scatter(df.Temperature,df.Humidity,color='red')
plt.title('Scatter Plot')
plt.xlabel('Temperature')
plt.ylabel('Humidity')
plt.show()


plt.figure(figsize=(6,4))
sns.heatmap(df[['Temperature','Humidity','WindSpeed','Rainfall']].corr(), annot=True, fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()



```
